# add correct pathways to the models

```python
from pandas import read_csv

pathways = read_csv("AGORA2_pathways.txt", sep="\t", encoding = "ISO-8859-1")
pathways
```

```python
from json import dump

rxnID_subsystem_dict = dict(zip(pathways["ReactionID"], pathways["Subsystem"]))
with open("rxnID_subsystem.json", "w") as jsonOut:   dump(rxnID_subsystem_dict, jsonOut, indent=3)
```

```python
pathways_rxnIDs = {}
for index, row in pathways.iterrows():
    if row["Subsystem"] in pathways_rxnIDs:   pathways_rxnIDs[row["Subsystem"]].append(row["ReactionID"])
    else:   pathways_rxnIDs[row["Subsystem"]] = [row["ReactionID"]]
```

# determining the set of excreted compounds

```python
from glob import glob
from pandas import read_csv

for fva in glob("AGORA2_fva/*"):
    fva_table = read_csv(fva)
    fva_table.index = fva_table["Unnamed: 0"]
    fva_table.drop(["Unnamed: 0"], axis=1)
    # fva_dict = dict(zip(fva_table["minimum"], fva_table.index))
    excreta = set([rxnId for rxnId, minFlux in fva_table["maximum"].to_dict().items() if minFlux < -0.01 and rxnId[:3] == "EX_"])
    # for rxn, row in fva_table.iterrows:
    #     if row["minimum"] <

excreted_mets = [rxnId.replace("EX_", "").replace("(e)", "") for rxnId in excreta]
```

```python
from json import load

with open("/Users/afreiburger/Documents/CommScores/commscores/data/MSDB_xRefs.json", 'r') as jsonIn:
    xRefs = load(jsonIn)

with open("/Users/afreiburger/Documents/CommScores/commscores/data/compoundNames.json", 'r') as jsonIn:
    nameCPD = load(jsonIn)

cpdName = {}
for k,v in nameCPD.items():
    if v not in cpdName:  cpdName[v] = k
    elif len(k) > len(cpdName[v]):   cpdName[v] = k

excreted_metNames = []
for metID in excreted_mets:
    try:  excreted_metNames.append(cpdName[xRefs["bigg"][metID]])
    except:
        try:  excreted_metNames.append(cpdName[xRefs["bigg"][metID.replace("_", "__")]])
        except:  print(metID)
```

```python
with open("excreta.txt", "w") as Out:
    Out.write("\n".join(excreted_metNames))
```
