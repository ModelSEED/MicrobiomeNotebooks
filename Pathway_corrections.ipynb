```python
from cobra.io import load_json_model, save_json_model, read_sbml_model, write_sbml_model
from glob import glob
from os import path, remove

def remove_groups(model_path):
    json_path = model_path.replace("xml", "json")
    try:
        save_json_model(read_sbml_model(model_path), json_path)
        write_sbml_model(load_json_model(json_path), model_path)
        remove(json_path)
    except:  pass
    

from datetime import datetime  ;  from multiprocess import Pool
pool = Pool(6)
args = [model_path for model_path in glob("AGORA2/*.xml") if path.getsize(model_path) > 1.08e7]
pool.map(remove_groups, args)
```
