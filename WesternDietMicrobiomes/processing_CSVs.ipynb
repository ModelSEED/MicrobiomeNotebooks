```python
from pandas import DataFrame
from numpy import log2

df_data = {}
trials = ['RC-ABX_-3', 'RC-ABX_0', 'RC-ABX_3', 'RC-ABX_5', 'RC-ABX_7', 'RC-ABX_11', 'RC-ABX_14', 'RC-ABX_28']
for i, diet_day in enumerate(trials):
    # if i == len(trials)-1:  break
    if diet_day not in sample_growth.index:
        print(f"Mayday;  missing {diet_day} in the Sample CSV")
        continue
    print(diet_day, end="\t+\t")
    future_trial = trials[i+1]
    if future_trial not in sample_growth.index:
        print(f"Mayday;  missing {future_trial} in the Sample CSV")
        continue
    print(future_trial, end="\t=\t")
    trial = diet_day.split("_")
    trial_name = trial[0]+f"_{(float(trial[1])+float(future_trial.split('_')[1]))/2}"
    print(trial_name)
    df_data[trial_name] = {column: max([-3, min([3, log2(sample_growth.loc[future_trial].values[index]/value)])])
                           for index, (column, value) in enumerate(sample_growth.loc[diet_day].items())}

changes_df = DataFrame(df_data).T

from json import load
with open("/Users/afreiburger/Documents/ModelSEEDpy/examples/Community Modeling/smetana/BiGG_to_MSID.json") as jsonOut:  biggMSID = load(jsonOut)
biggMSID
new_cols = []
for col in changes_df.columns:
    if col in biggMSID:   new_cols.append(biggMSID[col])
    else: new_cols.append(col) ; print(f"missing\t{col}")
changes_df.columns = new_cols

changes_df.index.name = "diet_day diff"
display(changes_df)
changes_df.to_csv("log2_fold_change_fluxes.csv")
```
