```python
from pandas import DataFrame
from numpy import log2

df_data = {}
trials = ['RC-ABX_-3', 'RC-ABX_0', 'RC-ABX_3', 'RC-ABX_5', 'RC-ABX_7', 'RC-ABX_11', 'RC-ABX_14', 'RC-ABX_28']
for i, diet_day in enumerate(trials):
    # if i == len(trials)-1:  break
    if diet_day not in sample_growth.index:
        print(f"Mayday;  missing {diet_day} in the Sample CSV")
        continue
    print(diet_day, end="\t+\t")
    future_trial = trials[i+1]
    if future_trial not in sample_growth.index:
        print(f"Mayday;  missing {future_trial} in the Sample CSV")
        continue
    print(future_trial, end="\t=\t")
    trial = diet_day.split("_")
    trial_name = trial[0]+f"_{(float(trial[1])+float(future_trial.split('_')[1]))/2}"
    print(trial_name)
    df_data[trial_name] = {column: max([-3, min([3, log2(sample_growth.loc[future_trial].values[index]/value)])])
                           for index, (column, value) in enumerate(sample_growth.loc[diet_day].items())}

changes_df = DataFrame(df_data).T

from json import load
with open("/Users/afreiburger/Documents/ModelSEEDpy/examples/Community Modeling/smetana/BiGG_to_MSID.json") as jsonOut:  biggMSID = load(jsonOut)
biggMSID
new_cols = []
for col in changes_df.columns:
    if col in biggMSID:   new_cols.append(biggMSID[col])
    else: new_cols.append(col) ; print(f"missing\t{col}")
changes_df.columns = new_cols

changes_df.index.name = "diet_day diff"
display(changes_df)
changes_df.to_csv("log2_fold_change_fluxes.csv")
```

## load and scalpt the metabolomics data

```python
# load the concentrations
ave_total_df = read_csv("../../metabolomics data/day_averaged_concentrations.csv")
# redefine the index
ave_total_df.reset_index()
ave_total_df.index = [f'{row["diet"]}_{row["Rec_day_adj"]}' for index, row in ave_total_df.iterrows()]
ave_total_df.drop(["Rec_day_adj", "diet"], axis=1, inplace=True)
ave_total_df.drop([index for index in ave_total_df.index if "PBS" in index], axis=0, inplace=True)
ave_total_df.index = [i.replace(".0", "") for i in ave_total_df.index]
# rename the columns
with open("/Users/afreiburger/Documents/ModelSEEDpy/examples/Community Modeling/smetana/BiGG_to_MSID.json") as jsonOut:  biggMSID = load(jsonOut)
biggMSID
new_cols = []
for col in ave_total_df.columns:
    if col in biggMSID:   new_cols.append(biggMSID[col])
    else: new_cols.append(col) ; print(f"missing\t{col}")
ave_total_df.columns = new_cols
display(ave_total_df)
```

## create the scatter plots

```python
from matplotlib import pyplot

# Create a single plot
pyplot.figure(figsize=(10, 8))  # Set the size of the figure

# Loop through each column pair
# colors = ['blue', 'green', 'red']  # Define a list of colors for differentiation
# markers = ['o', '^', 's']  # Different markers for each scatter plot

# remove sample rows that are not represented in the metabolomics data
sample_growth = read_csv("SamplegrowthInteractions.csv").set_index("Sample").drop("Name", axis=0).astype(float)
sample_growth_slimmed = sample_growth.drop([index for index in sample_growth.index if index in ave_total_df.index], axis=0)
print(sample_growth_slimmed.index)
print(ave_total_df.index)
sample_growth_slimmed = sample_growth_slimmed.loc[ave_total_df.index]
display(sample_growth_slimmed)

for col_x, col_y in zip(sample_growth_slimmed.columns, ave_total_df.columns):
    # display(sample_growth_slimmed[col_x])
    # print(sample_growth_slimmed[col_x].index)
    # print([x for x in sample_growth_slimmed[col_x].items()][0])
    # break
    xs = sample_growth_slimmed[col_x]  #[x for index, x in sample_growth_slimmed[col_x].items() if index in ]
    ys = ave_total_df[col_y]   #[ave_total_df[col_y]]
    pyplot.scatter(xs, ys, alpha=0.5, label=f'{col_x} vs {col_y}',
                # color=colors[i], 
                # marker=markers[i]
                )

# Adding labels and title
pyplot.xlabel('Computed')
pyplot.ylabel('Metabolomics')
pyplot.title('Correlation of computed versus experimental metabolic activity')
pyplot.legend()

# Show plot
pyplot.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

# create the datasets that will be fit into the model
x = np.array([1, 2, 3, 4, 5]).reshape((-1, 1))  # Make sure x is 2D
y = np.array([2, 4, 5, 4, 5])  # Dependent variable

# Fit the model
model.fit(x, y)

# Calculate R^2
r_squared = model.score(x, y)
print(f'Coefficient of Determination (R^2): {r_squared}')
```
