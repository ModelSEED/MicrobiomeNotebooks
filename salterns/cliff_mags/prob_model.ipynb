```python
genome = 'Salt_Pond_MetaGSF2_A_D1_MG_DASTool_bins_concoct_out.23.contigs__.RAST'
def fetch_annotation(genome, min_ani, db):
    annotation_self = set()
    annotation_other = set()
    for rast in db:
        gene_calls = db[rast].get(genome, {})
        if len(gene_calls) > 0:
            for gene_call, score in gene_calls.items():
                if score >= min_ani:
                    if gene_call.startswith('self:'):
                        annotation_self.add(rast)
                    else:
                        annotation_other.add(rast)
    return annotation_self, annotation_other
annotation_self, annotation_other = fetch_annotation(genome, 70, annotation_database_lo_85)
len(annotation_self), len(annotation_other), len(annotation_other & annotation_self), len(annotation_other - annotation_self)
```

```python
import math
sample = 'SF2_A_D1'
sample_data = relabund[sample]
total = 0
members = {}
round_figure = 4
for k in sample_data:
    if k != 'unbinned_reads':
        v = round(sample_data[k], round_figure)
        total += v
        if v != 0:
            #print(k, relabund[sample][k], v)
            members[k] = v
eps = 10000
total_p = 0
for k, v in members.items():
    vv = math.floor((v / total) * eps)
    total_p += vv
    #print(k, v / total, vv / eps)
total, len(members), total_p / eps
```
