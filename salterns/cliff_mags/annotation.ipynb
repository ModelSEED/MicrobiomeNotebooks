```python
template = kbase.get_from_ws('GramNegModelTemplateV5_test', 12218)
```

```python
from modelseedpy import MSBuilder
```

```python
feature_ct = 0

genome_annotation = MSGenome()
features = []
for a in annotation:
    ff = MSFeature(f'feature_{feature_ct}', "")
    ff.add_ontology_term('RAST', a)
    feature_ct += 1
    features.append(ff)
genome_annotation.add_features(features)
```

```python
b = MSBuilder(genome_annotation, template)
```

```python
model = b.build('super', annotate_with_rast=False)
```
